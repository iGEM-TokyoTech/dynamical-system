# 力学系

[[1](#footnote1)]より力学系とは以下の条件を満たす組 $(M, (g^t)_{t\in T})$ のことである．

1. $T$ が $\mathbb{Z},\mathbb{Z}^{+}, \mathbb{R}$ または $\mathbb{R}^+$ のいずれかであり， $t$ を期間， $T$ を時間集合と呼ぶ．
2. $M$ は空でない集合である． $\boldsymbol{x}\in M$ を状態， $M$ を状態空間と呼ぶ．
3. $(g^t)_{t\in T}$ は $M\rightarrow M$ なる関数の族で $T$ で添字付けされる． 任意の $t\in T$ において $g^t$ を期間 $t$ におけるシステムの状態遷移と呼ぶ．
4. 任意の $v, t \in T$ ， および任意の $\boldsymbol{x} \in M$ において $g^0(\boldsymbol{x}) =\boldsymbol{x}$， $g^{v+t}(\boldsymbol{x}) = g^{v}(g^{t}(\boldsymbol{x}))$

ここで，状態は時間が経つことで変化する( $v$ 経てば状態 $\boldsymbol{x}$ は $g^{v}(\boldsymbol{x})$ へと遷移する)が，システム自体 $(M, (g^t)_{t\in T})$ は変化しないことがわかる．

さらに，ある時点での状態を $\boldsymbol{x}_0$ とする．このとき，期間 $t$ 後の状態は $g^t(\boldsymbol{x}_0)$ と表せるがこれを $g^t(\boldsymbol{x}_0) = \boldsymbol{x}(t)$ とすれば
状態はある時点からの期間 **$t$ を独立変数とする関数 $\boldsymbol{x}(t)$ とすることができる** ．我々が扱う力学系における時間がこの $t$ にあたる．

力学系の本質は，時間変化に伴った状態変化の記述であって，ある時点における状態の記述ではない．


# 常微分方程式

> 常微分方程式（じょうびぶんほうていしき、英: ordinary differential equation, O.D.E.）とは、微分方程式の一種で、未知関数が本質的にただ一つの変数を持つものである場合をいう。 ([Wikiprdia より](https://ja.wikipedia.org/wiki/常微分方程式))

力学系は，時間の変化に伴った状態の変化を記述するものであった．状態を $k$ 個の変数で表す場合，ある時点での状態は $k$ 次元空間 (先述の状態空間 $M$ にあたる) の 1 点として表せ，時間 $t$ に伴う変化はその点の軌跡
( $\{g^{t}(\boldsymbol{x}) | \forall t \in T\}$ )として表現できる．
このとき， **$k$ 個の変数**は，上述の未知関数が $k$ 個あるもの，**時間 $t$** は $k$ 個の未知関数が本質的にただ一つ持っている変数にあたる．

> 未知関数が単独でない場合には、関数の組をベクトルの記法を用いて表せば次のようになる。
> $${\displaystyle {\boldsymbol {F}}(t,{\boldsymbol {x}}(t),{\boldsymbol {x}}^{(1)}(t),\dots ,{\boldsymbol {x}}^{(n)}(t))={\boldsymbol {0}}\quad \left({\boldsymbol {x}}^{(k)}(t):={\frac {\mathrm {d} ^{k}}{\mathrm {d} t^{k}}}{\boldsymbol {x}}(t),\,\mathrm {for} ~\,k=0,1,\dots ,n\right).}$$
> ここで $\boldsymbol {F}$, $\boldsymbol {x}$ は
> $${\displaystyle {\begin{aligned}&{\boldsymbol {F}}(t,{\boldsymbol {x}}(t),{\boldsymbol {x}}^{(1)}(t),\dots ,{\boldsymbol {x}}^{(n)}(t))=\left(F_{1}(t,{\boldsymbol {x}}(t),{\boldsymbol {x}}^{(1)}(t),\dots ,{\boldsymbol {x}}^{(n)}(t)),\dots ,F_{r}(t,{\boldsymbol {x}}(t),{\boldsymbol {x}}^{(1)}(t),\dots ,{\boldsymbol {x}}^{(n)}(t))\right),\\&{\boldsymbol {x}}(t)=\left(x_{1}(t),\dots ,x_{m}(t)\right)\end{aligned}}}$$
> を表す。この方程式系はしばしば連立常微分方程式と呼ばれる。 ([Wikiprdia より](https://ja.wikipedia.org/wiki/常微分方程式))

状態を $k$ 個の変数で表す場合，状態 $\boldsymbol {x}$ は $k$ 個の未知関数の組 $\left(x_{1}(t),\dots ,x_{k}(t)\right)$ に対応する．


**練習** <br>
時刻 $t$ に依存する未知関数によって表される $x(t)$ を変数とする状態を考える．$x(t)$ は自身の値に比例して増加することがわかっている．その比例定数を $r$ として常微分方程式を構成する．

$$
\displaystyle {\frac{d x}{d t} = rx}
$$

右辺を左辺に移行して

$$
\displaystyle {\frac{d x}{d t} - rx = 0}
$$

ここで以下のような $F(t, x(t), x^{(1)}(t))$ を考える．

$$
\displaystyle {F(t, x(t), x^{(1)}(t))} = x^{(1)}(t) - x(t)
$$

$x^{(1)}(t) = \frac{dx}{dt}$ より $F(t, x(t), x^{(1)}(t)) = \frac{d x}{d t} - rx = 0$ で常微分方程式が構成できた．

---

上述の微分方程式の解は $x(t) = x(0)e^{rt}$ と求まるが，せっかくなので具体的な初期値を設定し，数値計算でこれを求めてみる．(一般解が求められない場合は基本的に数値計算で解を見出す)

> 初期値問題（しょきちもんだい、英: Initial value problem）とは、未知関数のある点における値を初期条件として備えた常微分方程式を用いて、その未知変数の任意の点における値を求める問題のことを言う（コーシー問題とも呼ばれる）[Wikipedia より](https://ja.wikipedia.org/wiki/初期値問題)

$x(t_i) = x_i$ として，$t_0 = 0$，$x_0 = 1.0$ と初期値を設定する．$t_{i+1} - t_i = h\quad(i = 0,1,...)$ として，以下の方法で $x_{i+1}$ を求める．

$$
\displaystyle{x_{i+1} = x_{i} + h\frac{dx}{dt}(t_{i})}
$$

また，以下の方法でも求めることができる．

$$
\displaystyle{x_{i+1} = x_{i} + h\frac{dx}{dt}(t_{i+1})}
$$

後述の場合

$$
\displaystyle{x_{i+1} = x_{i} + h r x_{i+1}}
$$

より

$$
\displaystyle{x_{i+1} = \frac{x_{i}}{1-hr}}
$$

として求めることができる．


```python
import numpy as np
import matplotlib.pyplot as plt

r: float = 2.0  # r

t_0: float = 0.0

t_list: list[float] = [t_0]  # t のリスト． グラフの描画に必要

h: float = 0.01  # 微小時間 h

x_0: float = 10  # t=0 における x の初期値

exp_x_list: list[float] = [x_0]  # 求めたい未知関数 x の陽解法による解
imp_x_list: list[float] = [x_0]  # 求めたい未知関数 x の陰解法による解

t: float = t_0  # t
exp_x: float = x_0  # x_0
imp_x: float = x_0  # x_0

while t < 5.0:
    t += h
    t_list.append(t)

    exp_x += h * r * exp_x
    exp_x_list.append(exp_x)

    imp_x = imp_x / (1 - r * h)
    imp_x_list.append(imp_x)

# グラフの描画
plt.figure(figsize=(10, 6))
plt.plot(t_list, exp_x_list, "b", label="explicit curve")
plt.plot(t_list, imp_x_list, "y", label="implicit curve")
plt.plot(t_list, x_0 * np.exp(np.array(t_list) * r), "r--", label="$x_{0}e^{rt}$")
plt.title("Comparison of Simulated curve and $x_{0}e^{rt}$")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# SciPy の solve_ivp を利用すれば簡単に解くことができる
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

x0_0: float = (
    1.0  # ベクトル x の 1番目の要素の 初期値． ベクトル x は要素数が 1 なので実質実数である．
)
x_0: list[float] = [x0_0]  # solve_ivp 用に整形
r: float = 2.0  # 比例定数


def f(t: float, y: list[float]):
    """
    微分方程式
    dy[]/dt = f(t, y[]) なる f

    今回は dx/dt = rx より f(t, y[]) = [ry[0]]
    """
    return [r * y[0]]


t_start, t_end = 0, 5  # t の開始と終了
t_span = [t_start, t_end]  # solve_ivp 用に整形
t = np.linspace(t_start, t_end, 100)
solution = solve_ivp(f, t_span, x_0, t_eval=t)  # t_eval に評価したい t の
[x0] = solution.y  # unpack

# グラフの描画
plt.figure(figsize=(10, 6))
plt.plot(t, x0, "b", label="Simulated curve")
plt.plot(t, x_0[0] * np.exp(r * t), "r--", label="$x_{0}e^{rt}$")
plt.title("Comparison of Simulated curve and $x_{0}e^{rt}$")
plt.xlabel("t")
plt.ylabel("x")
plt.legend()
plt.grid(True)
plt.show()
```

# 力学系における常微分方程式

簡略化のため $f_i$ を以下のように定義する．

$$
{\displaystyle {\frac{d x_i}{d t} = f_i(x_1(t), x_2(t), \dots, x_k(t))}}
$$

例えば，ある時刻 $t$ における状態が状態空間上で $(x_1(t), x_2(t), \dots, x_k(t))$ と表せるとき，微小時間 $\varDelta t$ における，状態変数 $x_i$ の変化量は $f_i(x_1(t), x_2(t), \dots, x_k(t)) \varDelta t$ と表せる．

ここで， $f_i(x_1^{*}, x_2^{*}, \dots, x_k^{*})\quad (i = 1,2,\ldots,k)$ なる点 $(x_1^{*}, x_2^{*}, \dots, x_k^{*})$ を **固定点(平衡点)** とし $\delta x_i(t)$ を
$\delta x_i(t) = x_i(t) - x_i^{*}$ とする．

> 平衡点は、微分方程式の解を理解する上で重要で、平衡点を調べることは、微分方程式の解の定性的な振る舞いを知りたいときの最初の手段である。問題の微分方程式が非線形系の場合、解析的な解が得られることはまれだが、非線形系であっても平衡点を求めることなら可能である。[Wikipedia より](https://ja.wikipedia.org/wiki/平衡点)

上にもあるように，固定点を調べることは重要らしいので $f_j(x_1(t), x_2(t), \dots, x_k(t))$ に代入し， $f_j$ を固定点周りでテイラー展開する．
テイラー展開することで非線形なものも線形に落とし込めて分析しやすくなる．

<div style="">
<div style="border:solid 0.01rem gray; border-radius:0.5rem;width:95%; margin:auto;padding-left:1rem;padding-right:1rem;">
<b>多変数テイラー展開</b>

参考として $n$ 変数 $f(\boldsymbol{x})$ の $\boldsymbol{x} = \boldsymbol{a}$ 周りでのテイラー展開を以下に示す．

<div style="overflow-x:scroll;">

$$
\displaystyle {f(\boldsymbol{x}) = f(\boldsymbol{a}) + \sum_{i=1}^{n} \frac{\partial f}{\partial x_i}\bigg|_{\boldsymbol{a}}(x_i - a_i) + \frac{1}{2!} \sum_{i=1}^{n} \sum_{j=1}^{n} \frac{\partial^2 f}{\partial x_i \partial x_j}\bigg|_{\boldsymbol{a}}(x_i - a_i)(x_j - a_j) + \cdots + \frac{1}{k!} \sum_{i_1=1}^{n} \cdots \sum_{i_k=1}^{n} \frac{\partial^k f}{\partial x_{i_1} \cdots \partial x_{i_k}}\bigg|_{\boldsymbol{a}}(x_{i_1} - a_{i_1}) \cdots (x_{i_k} - a_{i_k}) + \cdots
}
$$

</div>
</div>
</div>

2 次以降の項は無視できるものとすれば以下の等式が成立する．

$$
\displaystyle {f_j(x_1(t), x_2(t), \dots, x_k(t)) = f_j(x_1^{*}, x_2^{*}, \dots, x_k^{*}) + \sum_i \delta x_i(t)\frac{\partial}{\partial x_i}f_j(x_1^{*}, x_2^{*}, \dots, x_k^{*})}
$$

さらに， $\delta x_i(t) = x_i(t) - x_i^{*}$ の両辺を $t$ で微分すると，

$$
\displaystyle{\frac{d \delta x_i(t)}{dt} = f_i(x_1(t), x_2(t), \dots, x_k(t))}
$$

ここで $(x_1^{*}, x_2^{*}, \dots, x_k^{*})$ は固定点より， $f_i(x_1^{*}, x_2^{*}, \dots, x_k^{*}) = 0$ であることに注意すると以下の微分方程式を得る．

$$
\displaystyle{\frac{d \delta x_j(t)}{dt} - \sum_i \delta x_i(t)\frac{\partial}{\partial x_i}f_j(x_1^{*}, x_2^{*}, \dots, x_k^{*}) = 0}
$$

上方程式が全ての $j = 1, 2\dots, k$ において成り立つので

$$
\displaystyle{\frac{d}{dt} \begin{pmatrix}
\delta x_1(t) \\
\delta x_2(t) \\
\vdots \\
\delta x_k(t)
\end{pmatrix}
-
\begin{pmatrix}
\frac{\partial}{\partial x_1} f_1(\boldsymbol{x}^{*}) & \frac{\partial}{\partial x_2} f_1(\boldsymbol{x}^{*}) & \cdots & \frac{\partial}{\partial x_k} f_1(\boldsymbol{x}^{*}) \\
\frac{\partial}{\partial x_1} f_2(\boldsymbol{x}^{*}) & \frac{\partial}{\partial x_2} f_2(\boldsymbol{x}^{*}) & \cdots & \frac{\partial}{\partial x_k} f_2(\boldsymbol{x}^{*}) \\
\vdots & \vdots & \ddots & \vdots \\
\frac{\partial}{\partial x_1} f_k(\boldsymbol{x}^{*}) & \frac{\partial}{\partial x_2} f_k(\boldsymbol{x}^{*}) & \cdots & \frac{\partial}{\partial x_k} f_k(\boldsymbol{x}^{*})
\end{pmatrix}
\begin{pmatrix}
\delta x_1(t) \\
\delta x_2(t) \\
\vdots \\
\delta x_k(t)
\end{pmatrix}
= \begin{pmatrix}
0 \\
0 \\
\vdots \\
0
\end{pmatrix}}
$$

($(x_1^{*}, x_2^{*}, \dots, x_k^{*}) = \boldsymbol{x}^{*}$ とした)

ベクトル表記 $\boldsymbol{\delta x} = \begin{pmatrix}\delta x_1(t) \\\delta x_2(t) \\\vdots \\\delta x_k(t)\end{pmatrix}$,
$\boldsymbol{f} = \begin{pmatrix}f_1 \\f_2 \\\vdots \\ f_k\end{pmatrix}$，
$D_x = \begin{pmatrix}\frac{\partial}{\partial x_1} \\\frac{\partial}{\partial x_2} \\\vdots \\\frac{\partial}{\partial x_k} \\\end{pmatrix}$ を用いれば

$$
\displaystyle{\frac{d \boldsymbol{\delta x}}{dt} - \left(D_x\boldsymbol{f}\right)^\top|_{\boldsymbol{x}^*}\boldsymbol{\delta x} = \boldsymbol{0} }
$$

と表せる．
これは $\boldsymbol{\delta x}$ およびその一階微分である $\frac{d \boldsymbol{\delta x}}{dt}$ からなる方程式であり，線形の常微分方程式となる．

上のような行列 $\boldsymbol{f} D_x^\top$ を $\boldsymbol{f}$ のヤコビ行列といい $J$ と書くことにする．


---


**練習**<br>

> ロトカ・ヴォルテラの方程式（ロトカ・ヴォルテラのほうていしき、英語: Lotka-Volterra equations）とは、生物の捕食-被食関係による個体数の変動を表現する数理モデルの一種。2 種の個体群が存在し、片方が捕食者、もう片方が被食者のとき、それぞれの個体数増殖速度を二元連立非線形常微分方程式系で表現する。ロトカ・ヴォルテラの捕食式やロトカ・ヴォルテラ捕食系、ロトカ-ヴォルテラの捕食者-被食者モデルなどとも呼ばれる [Wikipedia より](https://ja.wikipedia.org/wiki/ロトカ・ヴォルテラの方程式)

具体的には以下の非線形の一階微分方程式で表される．( $xy$ の項 )

$$
{\displaystyle {\begin{aligned}{\frac {dx}{dt}}&=ax-bxy,\\ {\frac {dy}{dt}}&=cxy-dy.\end{aligned}}}
$$

ここで $t$ は時間，$x$ は被食者の個体数， $y$ は捕食者の個体数で共に $t$ を実質的なただ一つの変数とする関数で表せるものとする．
$a,b,c,d$ は正の実数である．

ここで $\boldsymbol{f} =\begin{pmatrix}\frac{dx}{dt}\\ \frac{dy}{dt}\end{pmatrix}$ とすれば $\boldsymbol{f}$ のヤコビ行列 $J$ は

$$
\begin{aligned}
J
&= \begin{pmatrix}\frac{\partial}{\partial x}\frac{dx}{dt} & \frac{\partial}{\partial y}\frac{dx}{dt}\\ \frac{\partial}{\partial x}\frac{dy}{dt} & \frac{\partial}{\partial y}\frac{dy}{dt} \end{pmatrix} \\
&=  \begin{pmatrix}a-by & -bx\\ cy & cx - d\end{pmatrix}
\end{aligned}
$$

平衡点は以下の連立方程式の解である．

$$
\displaystyle{\left\{\begin{aligned}x(a-by) = 0\\ y(cx-d) = 0\end{aligned}\right.}
$$

よって $(x,y) = (0, 0), (\frac{d}{c}, \frac{a}{b})$ が平衡点である．


```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 定数
a: float = 3.0
b: float = 1.5
c: float = 1.0
d: float = 2.0

x0_stars = [0.0, d / c]  # 平衡点の x^{*} x 座標 (1つ目の要素たち)
x1_stars = [0.0, a / b]  # 平衡点の x^{*} y 座標 (2つ目の要素たち)
x_stars = np.array([x0_stars, x1_stars])  # 平衡点 x^{*}

x0_0: float = 4.0  # 被食者 x の初期値
x1_0: float = 1.0  # 捕食者 y の初期値
x_0: np.ndarray = np.array([x0_0, x1_0], dtype=np.float64)  # solve_ivp 用に整形


def f(t: float, y: np.ndarray):
    """
    微分方程式
    dy[]/dt = f(t, y[]) なる f

    今回は dx/dt = ax - bxy， dy/dt = cxy - dy

    x = y[0]，y = y[1]

    とすれば f(t, y[]) = [ay[0] - by[0]y[1], cy[0]y[1] - dy[1]]
    """
    return np.array([a * y[0] - b * y[0] * y[1], c * y[0] * y[1] - d * y[1]])


t_start, t_end = 0, 20  # t は 0 ~ 20
t_span = [t_start, t_end]

t = np.linspace(t_start, t_end, 100)  # f を評価する t
solution = solve_ivp(f, t_span, x_0, t_eval=t)
[x0, x1] = solution.y  # unpack

# グラフの描画
plt.figure(figsize=(10, 6))

plt.plot(t, x0, "b", label="prey")
plt.plot(t, x1, "y", label="predator")
plt.title("prey and predator vs time")
plt.xlabel("time")
plt.ylabel("population")
plt.legend()
plt.grid(True)
plt.show()
```

ここで，平衡点周りでの $\frac{dx}{dt}$，$\frac{dy}{dt}$ の動きを見てみる．


```python
x_points = np.linspace(x_stars[0][0], 2 * x_stars[0][1], 8)  # f に渡すx座標
y_points = np.linspace(x_stars[1][0], 2 * x_stars[1][1], 8)  # f に渡すy座標

# x, y座標の組み合わせを生成
X, Y = np.meshgrid(x_points, y_points)

# ベクトル場の計算
U, V = f(0.0, np.array([X, Y]))

# プロット
plt.figure(figsize=(8, 6))
plt.title("vector field around equilibrium points")
plt.quiver(X, Y, U, V, label="vector field")
plt.plot(
    x_stars[0], x_stars[1], linestyle="None", marker="o", label="equilibrium points"
)
plt.legend()
plt.show()
```
