**エノン写像**

以下のような $\mathbb{R}^2$ 上の関数 $\boldsymbol{f}$ をエノン写像という．

$$
\boldsymbol{f}(x, y) = (a - x^2 + by, x)
$$

</div>
</div>

$a = 1.28$， $b = -0.3$ で初期値が $(0, 0)$ と $(-2.5, 0)$ とした場合をシミュレーションするコードを以下に示す．


```python
import matplotlib.pyplot as plt


def henon_map_gen(a: float, b: float):
    def henon_map(x: float, y: float):
        return (a - x * x + b * y, x)

    return henon_map


f = henon_map_gen(1.28, -0.3)

x_1_init, y_1_init = 0, 0
x_2_init, y_2_init = -2.5, 0

X_1: list[float] = [x_1_init]
Y_1: list[float] = [y_1_init]
X_2: list[float] = [x_2_init]
Y_2: list[float] = [y_2_init]

steps = 30

for _ in range(steps):
    x_1, y_1 = f(X_1[-1], Y_1[-1])
    x_2, y_2 = f(X_2[-1], Y_2[-1])
    X_1.append(x_1)
    Y_1.append(y_1)
    X_2.append(x_2)
    Y_2.append(y_2)

fig, axs = plt.subplots(2, 1, figsize=(8, 12))

axs[0].scatter(X_1, Y_1, color="b", label="Starting from (0, 0)")
axs[0].set_title("Henon Map - Starting from (0, 0)")
axs[0].set_xlabel("X")
axs[0].set_ylabel("Y")
axs[0].legend()
axs[0].grid(True)

axs[1].scatter(X_2, Y_2, color="r", label="Starting from (-2.5, 0)")
axs[1].set_title("Henon Map - Starting from (-2.5, 0)")
axs[1].set_xlabel("X")
axs[1].set_ylabel("Y")
axs[1].legend()
axs[1].grid(True)

plt.tight_layout()
plt.show()
```

$(0, 0)$ から始めた場合，2 点に吸引されているのに対し， $(-2.5, 0)$ では無限に発散している．
このように，初期値によって吸引される領域は変わってくることがわかる．
