# ヤコビ行列と平衡点

$$
\displaystyle{\frac{d \boldsymbol{\delta x}}{dt} = J|_{\boldsymbol{x}^*}\boldsymbol{\delta x}}
$$

平衡点周りの状態 $\boldsymbol{x} = \boldsymbol{x}^* + \boldsymbol{\delta x}$ の時間変化 $\frac{d \boldsymbol{\delta x}}{dt}$ は上のように近似できた．任意の $\boldsymbol{\delta x}$ において上式が成り立つとすれば(?)
平衡点周りの状態の動きは $\boldsymbol{\delta x}$ を直接見ずとも， $J|_{\boldsymbol{x}^*}$ を見ればその特徴を理解できそうである．

ここでは， $J$ を対角化することで状態の平衡点周りでの動きを探ってみる．

<b>固有値</b>

> 線形空間 $V$（有限次元とは限らない）上の線形変換 $A$ に対して、次の方程式
> $${\displaystyle A{\boldsymbol {x}}=\lambda {\boldsymbol {x}}}$$
> を満たす零ベクトルでないベクトル $\boldsymbol{x}$ とスカラー $\lambda$ が存在するとき、$\boldsymbol{x}$ を $A$ の固有ベクトル（右固有ベクトル）、$\lambda$ を $A$ の固有値と呼ぶ。[Wikipedia より](https://ja.wikipedia.org/wiki/固有値と固有ベクトル)

例えば $\begin{pmatrix}1 & 2\\ 1 & 0\end{pmatrix}$ は $\begin{pmatrix}1 & 2\\ 1 & 0\end{pmatrix}\begin{pmatrix}2t\\ t\end{pmatrix} = 2\begin{pmatrix}2t\\ t\end{pmatrix}$，
$\begin{pmatrix}1 & 2\\ 1 & 0\end{pmatrix}\begin{pmatrix}t\\ -t\end{pmatrix} = -1\begin{pmatrix}t\\ -t\end{pmatrix}$ で固有値は $2, -1$，固有ベクトルは
$\begin{pmatrix}2t\\ t\end{pmatrix}\begin{pmatrix}t\\ -t\end{pmatrix}$ である．

</div>
<br/>
<div style="border:solid 0.01rem gray; border-radius:0.5rem;width:95%; margin:auto;padding-left:1rem;padding-right:1rem;">
<b>対角化</b>

対角成分以外が $0$ である行列を[対角行列](https://ja.wikipedia.org/wiki/対角行列)と呼ぶ．
${\displaystyle {\begin{bmatrix}c_{1}&&&0\\&c_{2}&&\\&&\ddots &\\0&&&c_{n}\end{bmatrix}}}$ のようなもののことである．

> 対角化（たいかくか、diagonalization[1]）とは、正方行列を適当な線形変換によりもとの行列と相似な対角行列に変形することを言う。[Wikipedia より](https://ja.wikipedia.org/wiki/対角化)

先ほどの固有値の知識を用いて対角化が可能である． $k\times k$ 正方行列 $A$ に対する 固有値を $\lambda_1, \lambda_2,\dots,\lambda_k$，それに対する固有ベクトルを $\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k$ とする．ここで，以下の式が成り立つ．

$$
\displaystyle{A\begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix} = \begin{pmatrix}\lambda_1\boldsymbol{x}_1,\lambda_2\boldsymbol{x}_2,\dots,\lambda_k\boldsymbol{x}_k\end{pmatrix}}
$$

さらに $\begin{pmatrix}\lambda_1\boldsymbol{x}_1,\lambda_2\boldsymbol{x}_2,\dots,\lambda_k\boldsymbol{x}_k\end{pmatrix} = \begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix} \begin{pmatrix}\lambda_{1}&&&0\\&\lambda_{2}&&\\&&\ddots &\\0&&&\lambda_{k}\end{pmatrix}$ より

$$
\displaystyle{A\begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix} = \begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix} \begin{pmatrix}\lambda_{1}&&&0\\&\lambda_{2}&&\\&&\ddots &\\0&&&\lambda_{k}\end{pmatrix}}
$$

両辺に左から $\begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix}$ の逆行列 $\begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix}^{-1}$ をかければ

$$
\displaystyle{\begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix}^{-1} A \begin{pmatrix}\boldsymbol{x}_1,\boldsymbol{x}_2,\dots,\boldsymbol{x}_k\end{pmatrix} = E\begin{pmatrix}\lambda_{1}&&&0\\&\lambda_{2}&&\\&&\ddots &\\0&&&\lambda_{k}\end{pmatrix} = \begin{pmatrix}\lambda_{1}&&&0\\&\lambda_{2}&&\\&&\ddots &\\0&&&\lambda_{k}\end{pmatrix}}
$$

平衡点におけるヤコビ行列 $J|_{\boldsymbol{x}^*}$ を対角化することを考える． 対角化された対角行列を $D$ として $D = P^{-1}J|_{\boldsymbol{x}^*}P$ と対角化する．
ここで

$$
\displaystyle{P^{-1}\frac{d \boldsymbol{\delta x}}{dt} =P^{-1}J|_{\boldsymbol{x}^*}\boldsymbol{\delta x}}
$$

さらに $J|_{\boldsymbol{x}^*}$ と $\boldsymbol{\delta x}$ の間に $E = PP^{-1}$ を入れれば

$$
\displaystyle{P^{-1}\frac{d \boldsymbol{\delta x}}{dt} = D P^{-1}\boldsymbol{\delta x}}
$$

$P^{-1}\frac{d \boldsymbol{\delta x}}{dt} =\frac{d P^{-1}\boldsymbol{\delta x}}{dt} $， $\boldsymbol{v} = P^{-1}\boldsymbol{\delta x}$ とすれば

$$
\displaystyle{\frac{\boldsymbol{d \boldsymbol{v}}}{dt} = D\boldsymbol{v}}
$$

よって $v_i(t) = v_i(0)e^{\lambda_i t}$．

さらに $\displaystyle{v_i = \sum_mP^{-1}_{im}\delta x_{m}}$， $\displaystyle{\delta x_i = \sum_m P_{im}v_m}$ より

$$
\displaystyle{v_i(t) = e^{\lambda_i t}\sum_m P^{-1}_{im}\delta x_{m}(0)}
$$

よって，

$$
\displaystyle{\delta x_i(t) = \sum_m P_{im} \left(e^{\lambda_m t}\sum_j P^{-1}_{mj}\delta x_{j}(0)\right) = \sum_{m, j}e^{\lambda_m t}P_{im}P^{-1}_{mj}\delta x_{j}(0)}
$$

虚数単位 $i$ を用いて $\lambda_i = a + \omega i \quad a, \omega \in \mathbb{R}$ とすれば $e^{\lambda_i t} = e^{a_it}(\cos\omega_i t + i\sin\omega_i t)$ より

$$
\displaystyle{\delta x_i(t) = \sum_{m, j} e^{a_mt}(\cos\omega_m t + i\sin\omega_m t)P_{im}P^{-1}_{mj}\delta x_{j}(0)}
$$

となる．


---


**練習**<br/>
`ode.ipynb` で扱った

$$
{\displaystyle {\begin{aligned}{\frac {dx}{dt}}&=ax-bxy,\\ {\frac {dy}{dt}}&=cxy-dy.\end{aligned}}}
$$

でこれを試してみよう．

$$
\displaystyle{ J = \begin{pmatrix}a-by & -bx\\ cy & cx - d\end{pmatrix}}
$$

$(0, 0)$ においては $J = \begin{pmatrix}a & 0\\ 0 & - d\end{pmatrix}$．固有値は $\lambda_{1} = a, \lambda_{2}=-d$ ，変換行列の一つとして $P=P^{-1}=E$ が求められる．

よって

$$
\displaystyle{\begin{aligned}\delta x_1(t) &=\sum_{m=1}^{2} \left(e^{a_mt}(\cos\omega_m t + i\sin\omega_m t)P_{1m}\sum_{j=1}^{2}P^{-1}_{mj}\delta x_{j}(0)\right)\\&=e^{at}\delta x_1(0)\end{aligned}}
$$

$$
\displaystyle{\begin{aligned}\delta x_2(t) &=\sum_{m=1}^{2} \left(e^{a_mt}(\cos\omega_m t + i\sin\omega_m t)P_{2m}\sum_{j=1}^{2}P^{-1}_{mj}\delta x_{j}(0)\right)\\&=e^{-dt}\delta x_2(0)\end{aligned}}
$$

$(\frac{d}{c}, \frac{a}{b})$ においては $J = \begin{pmatrix}0 & -\frac{bd}{c}\\ \frac{ac}{b} & 0\end{pmatrix}$．固有値は $\lambda_1=\sqrt{ad}i, \lambda_2=-\sqrt{ad}i$ ，
変換行列の一つとして $P=\begin{pmatrix}\frac{b}{c} &\frac{b}{c}\\-\sqrt{\frac{a}{d}}i & \sqrt{\frac{a}{d}}i\end{pmatrix}$ ，
$P^{-1} = \begin{pmatrix}\frac{c}{2b} & \frac{1}{2}\sqrt{\frac{d}{a}}i \\ \frac{c}{2b} & -\frac{1}{2}\sqrt{\frac{d}{a}}i\end{pmatrix}$ が求められる．

よって

$$
\displaystyle{\begin{aligned}\delta x_1(t) &=\sum_{m=1}^{2} \left(e^{a_mt}(\cos\omega_m t + i\sin\omega_m t)P_{1m}\sum_{j=1}^{2}P^{-1}_{mj}\delta x_{j}(0)\right)\\
&=\delta x_1(0)\cos(\sqrt{ad}t) - \frac{b}{c}\sqrt{\frac{d}{a}}\delta x_2(0)\sin(\sqrt{ad}t)\end{aligned}}
$$

$$
\displaystyle{\begin{aligned}\delta x_2(t) &=\sum_{m=1}^{2} \left(e^{a_mt}(\cos\omega_m t + i\sin\omega_m t)P_{2m}\sum_{j=1}^{2}P^{-1}_{mj}\delta x_{j}(0)\right)\\
&=\delta x_2(0)\cos(\sqrt{ad}t) + \frac{c}{b}\sqrt{\frac{a}{d}}\delta x_1(0)\sin(\sqrt{ad}t)\end{aligned}}
$$

それでは上の式が実際に平衡点周りでの動きを近似できてるか確かめてみる．


```python
import numpy as np
import matplotlib.pyplot as plt

# 定数
a: float = 3.0
b: float = 1.5
c: float = 1.0
d: float = 2.0

x0_stars = [0.0, d / c]  # 平衡点の x^{*} x 座標 (1つ目の要素たち)
x1_stars = [0.0, a / b]  # 平衡点の x^{*} y 座標 (2つ目の要素たち)
x_stars = np.array([x0_stars, x1_stars])  # 平衡点 x^{*}

t_start, t_end = 0, 20  # t は 0 ~ 20
t_span = [t_start, t_end]
t = np.linspace(t_start, t_end, 100)  # 評価する t


###
# ここから delta x を計算するプログラム
###


x0_0: float = 0.1  # 被食者 x の初期値 0, 0 周り
x1_0: float = 2.1  # 捕食者 y の初期値 0, 0 周り


def deltay0_1(t: np.ndarray):
    """
    (0, 0) 周りでの delta x_1
    """
    return np.exp(a * t) * (x0_0 - x_stars[0][0])


def deltay1_1(t: np.ndarray):
    """
    (0, 0) 周りでの delta x_2
    """
    return np.exp(-1 * d * t) * (x1_0 - x_stars[1][0])


deltax_1_plus_star = [
    deltay0_1(t) + x_stars[0][0],
    deltay1_1(t) + x_stars[1][0],
]  # 求めたのは \delta x で 平衡点からのズレを求めているので，平衡点の座標を加えて直す


x0_0: float = d / c * 1.1  # 被食者 x の初期値 d/c, a/b 周り
x1_0: float = a / b * 1.1  # 捕食者 y の初期値 d/c, a/b 周り


def deltay0_2(t: np.ndarray):
    """
    (d/c, a/b) 周りでの delta x_1
    """
    return (x0_0 - x_stars[0][1]) * np.cos(np.sqrt(a * d) * t) - (d / c) * np.sqrt(
        d / a
    ) * (x1_0 - x_stars[1][1]) * np.sin(np.sqrt(a * d) * t)


def deltay1_2(t: np.ndarray):
    """
    (d/c, a/b) 周りでの delta x_2
    """
    return (x1_0 - x_stars[1][1]) * np.cos(np.sqrt(a * d) * t) + (c / b) * np.sqrt(
        a / d
    ) * (x0_0 - x_stars[0][1]) * np.sin(np.sqrt(a * d) * t)


deltax_2_plus_star = [deltay0_2(t) + x_stars[0][1], deltay1_2(t) + x_stars[1][1]]


###
# ここからは状態 x のベクトル場を計算するプログラム
###


def f(t: float, y: np.ndarray):
    """
    微分方程式
    dy[]/dt = f(t, y[]) なる f

    今回は dx/dt = ax - bxy， dy/dt = cxy - dy

    x = y[0]，y = y[1]

    とすれば f(t, y[]) = [ay[0] - by[0]y[1], cy[0]y[1] - dy[1]]
    """
    return np.array([a * y[0] - b * y[0] * y[1], c * y[0] * y[1] - d * y[1]])


x_points = np.linspace(x_stars[0][0], 2 * x_stars[0][1], 8)  # f に渡すx座標
y_points = np.linspace(x_stars[1][0], 2 * x_stars[1][1], 8)  # f に渡すy座標

# x, y座標の組み合わせ(グリッド)を生成
X, Y = np.meshgrid(x_points, y_points)

# ベクトル場の計算
U, V = f(0.0, np.array([X, Y]))

###
# ここから描画のプログラム
###

# プロット
plt.figure(figsize=(8, 6))
plt.title("vector field around equilibrium points")
plt.xlim(np.min(X) - 0.5, np.max(X) * 1.1)
plt.ylim(np.min(Y) - 0.5, np.max(Y) * 1.1)
plt.quiver(X, Y, U, V, label="vector field")
plt.plot(
    x_stars[0], x_stars[1], linestyle="None", marker="o", label="equilibrium points"
)
plt.plot(
    deltax_1_plus_star[0],
    deltax_1_plus_star[1],
    "g",
    label="approximation around (0, 0)",
)
plt.plot(
    deltax_2_plus_star[0],
    deltax_2_plus_star[1],
    "r",
    label="approximation around (d/c, a/b)",
)
plt.legend()
plt.show()

plt.show()
```

状態空間のベクトル場とヤコビ行列を用いて得られる平衡点周りの状態の奇跡が重なることがわかる．
つまり $J|_{\boldsymbol{x}^*}$ の対角化によって平衡点周りの状態変化を数式として表すことができたことになる．

ここで $J|_{\boldsymbol{x}^*}$ と平衡点周りでの状態変化を簡単に解説する．

各状態変数の動きは $e^{(a+i\omega)t} = e^{a}(\cos\omega t + i\sin\omega t)$ の重ね合わせで表現できたわけだが， $e^{at}$ は平衡点からの距離(半径)に関する変化，
$(\cos\omega t + i\sin\omega t)$ は平衡点を中心とした回転変化を表す項と捉えてほしい．

つまり $a > 0$ ならその変化は時間が進むごとに増幅していく．
対して， $a < 0$ ならその変化は減少し，ある点に落ち着く．
$\omega = 0$ なら回転せず，そうでなければ回転する，というイメージである．

$(\frac{d}{c}, \frac{a}{b})$ 周りでの近似がどの点にも落ち着いてないのは $a = 0$ で一定の半径を保ったまま回転しているためである．
